# DISPLAY SEASON WINNER

```python
# Season winners for first 50

# Groupby() function is used to group datapoints based on a specific features that 
# facilitating the analysis and aggregation of information within each group.

season_winner = df.groupby('season')['winner'].value_counts().head(50)
print(season_winner)
```

# VISUALIZING NUMBER OF MATCHES PLAYED IN EACH IPL SEASON

```python
plt.figure(figsize=(15, 7))
sns.countplot(x = 'season', data = df, palette = 'summer')
plt.xticks(rotation = 90)
plt.xlabel('Season', fontsize = 15)
plt.ylabel('Matches', fontsize = 15)
plt.title('Number of matches played in each IPL season', fontsize = 15)
plt.show()
```

# You can see the above plot, that the number of matches played from 2011 to 2013 is above the range 70, in comparison from rest of the season 2013 is the highest matches played in an individual season.

```python
plt.figure(figsize=(15, 6))
sns.countplot(x = 'winner', data = df, palette = 'husl')
plt.xticks(rotation = 80)
plt.xlabel('Teams', fontsize = 15)
plt.ylabel('No. of Wins', fontsize = 15)
plt.title('Number of matches played by teams', fontsize = 15)
plt.show()
```

```python
df['win_by'] = np.where(df['win_by_runs'] > 0, 'Bat first', 'Field first')
```

# DISPLAY THE MATCH RESULTS USING PIE PLOT

```python
win = df.win_by.value_counts()
labels = np.array(win.index)
sizes = win.values
color = ['y','r']
plt.figure(figsize = (15, 8))
plt.pie(sizes, labels = labels, colors = color, autopct = '%1.1f%%', startangle = 90)
plt.axis('equal')
plt.title('Result Of The Match', fontsize = 20)
plt.show()
```

# NO. OF MATCHES WON BY BATTING AND FIELDING USING COUNTPLOT

```python
plt.figure(figsize=(15, 6))
sns.countplot(x = 'season', hue = 'win_by', data = df, palette = 'husl')
plt.xticks(rotation = 80)
plt.xlabel('Season', fontsize = 15)
plt.ylabel('Result', fontsize = 15)
plt.title('Number of matches won by Batting and Fielding First', fontsize = 15)
plt.show()
```

# In the Above Plot we can clearly see that the domination of Fielding First is more than Batting, except 2010 and 2015 rest all seasons most of the player choose to Field First and their winning result in all season is above 25, we can see that 2011 and 2015 the winning ratio is similar compare to other season.

# DISPLAY THE TOSS DECISION USING PIE CHART

```python
Toss = df.toss_decision.value_counts()
labels = np.array(Toss.index)
sizes = Toss.values
color = ['y','g']
plt.figure(figsize = (15, 8))
plt.pie(sizes, labels = labels, colors = color, autopct = '%1.1f%%', startangle = 90)
plt.axis('equal')
plt.title('Toss Result', fontsize = 20)
plt.show()
```

# NUMBER OF FOURS HIT BY TEAMS

```python
hit_fours = df2[df2['batsman_runs'] == 4]
hit_fours.groupby('batting_team')['batsman_runs'].agg([('runs_by_fours', 'sum'),('fours', 'count')]).sort_values('fours', ascending = False)
```

# PLOTTING A GRAPH ON FOURS(Boundary) HIT BY PLAYERS

```python
Batsman_fours = hit_fours.groupby('batsman')['batsman_runs'].agg([('fours', 'count')]).reset_index().sort_values('fours', ascending = False)
ax = Batsman_fours.iloc[:10, :].plot('batsman' ,'fours', kind = 'bar', color = 'orange', figsize = (15, 7))
plt.xticks(rotation = 'horizontal')
plt.xlabel('Player Names', fontsize = 15)
plt.ylabel('No. of Fours', fontsize = 15)
plt.title('Number of Fours hit by Players', fontsize = 20)
plt.show()
```

# NUMBER OF SIXES HIT BY TEAMS

```python
hit_sixes = df2[df2['batsman_runs'] == 6]
hit_sixes.groupby('batting_team')['batsman_runs'].agg([('runs_by_sixes', 'sum'),('sixes', 'count')]).sort_values('sixes', ascending = False)
```

# PLOTTING A GRAPH ON SIXES(Boundary) HIT BY AN INDIVIDUAL PLAYER

```python
Batsman_sixes = hit_sixes.groupby('batsman')['batsman_runs'].agg([('sixes', 'count')]).reset_index().sort_values('sixes', ascending = False)
ax = Batsman_sixes.iloc[:10, :].plot('batsman' ,'sixes', kind = 'bar', color = 'lightblue', figsize = (15, 7))
plt.xticks(rotation = 'horizontal')
plt.xlabel('Player Names', fontsize = 15)
plt.ylabel('No. of Sixes', fontsize = 15)
plt.title('Number of Sixes hit by Players', fontsize = 20)
plt.show()
```

# TOP 10 LEADING RUN SCORER IN IPL SEASON

```python
Batsman_score = df2.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum', ascending = False)
Batsman_score = Batsman_score.rename(columns = {'sum' : 'batsman_runs'})
print('**The Top 10 Leading Run Scorer**')
Batsman_score.iloc[:10] 
```

# TOP 10 BATSMAN WHO PLAYED THE MOST MATCHES

```python
No_matches_by_Player = df2[['match_id', 'player_dismissed']]
No_matches_by_Player = No_matches_by_Player.groupby('player_dismissed')['match_id'].count().reset_index().sort_values(by = 'match_id', ascending = False)
No_matches_by_Player.columns = ['batsman', 'No_of_Matches']
No_matches_by_Player[:10]
```
